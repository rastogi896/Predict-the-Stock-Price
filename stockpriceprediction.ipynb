```python
train=train.dropna()
```

```python
train.head()
```

```python
y_train=train['Price']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['type_of_shot'] = le.fit_transform(train.type_of_shot.values)
```

```python
test['Company'] = le.fit_transform(test.Company.values)
```

```python
test_id=test['ID']
```

```python
train.drop(['ID','Date','Price'],axis=1,inplace=True)
test.drop(['ID','Date'],axis=1,inplace=True)
```

```python
len(test)
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score
EPOCHS = 16
kf = KFold(n_splits = EPOCHS, shuffle = True)
y_preds = np.zeros(len(test))
y_oof = np.zeros(train.shape[0])
for tr_idx, val_idx in kf.split(train, y_train):
    clf = LinearRegression()   
    X_tr, X_vl = train.iloc[tr_idx, :], train.iloc[val_idx, :]
    y_tr, y_vl = y_train.iloc[tr_idx], y_train.iloc[val_idx]
    clf.fit(X_tr, y_tr)
    y_pred_train = clf.predict(X_vl)
    y_oof[val_idx] = y_pred_train
    #print('ROC AUC {}'.format(roc_auc_score(y_vl, y_pred_train)))    
    y_preds+= clf.predict(test) / EPOCHS
```

```python
y_preds
```

```python
y_preds=y_preds.tolist()
```

```python
columns = ['Price']
df1 = pd.DataFrame(data=y_preds, columns=columns)
df1['ID'] = test_id
df1 = df1[['ID','Price']]
```
